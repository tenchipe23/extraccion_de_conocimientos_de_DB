```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("test.csv")
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ["cabin"], inplace=True)
```

```python
df.columns
```

```python
df["age"]= df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
df["fare"]= df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["age"].min()
```

```python
df["age"].max()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(df, x="sex", hue="pclass")
plt.title("Total de genero por clase")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x="sex", y="age", hue="pclass")
plt.title("Rango de edad por genero y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```
