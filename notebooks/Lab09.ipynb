```python
from faker import Faker
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
fake = Faker('es_MX')
```

```python
datos = []
for i in range(1000):
    datos.append({
        'id_usuario': i+1,
        'nombre': fake.name(),
        'correo': fake.email(),
        'edad': random.randint(18,65),
        'seguidores': random.randint(100,5000),
        'platform': random.choice(
            ['TikTok', 'Hinstagram', 'Youtube']
        )
    })
```

```python
df = pd.DataFrame(datos)
df.info()
```

```python
df.describe(include="all")
```

```python
plt.figure(figsize=(8,4))
sns.histplot(
    df["edad"],
    bins=10
)
plt.title("Distribucion de edades")
plt.show()
```

```python
df_sucio =  df.copy()
```

```python
# Colocar valores nulos
for i in random.sample(range(len(df_sucio)), 5):
    df_sucio.loc[i, "nombre"] = None
```

```python
df_sucio.info()
```

```python
for i in random.sample(range(len(df_sucio)),5):
    df_sucio.loc[i,'edad'] = -10
```

```python
df_sucio.describe()['edad']
```

```python
for i in random.sample(range(len(df_sucio)),5):
    df_sucio.loc[i,'seguidores'] = 5000000
```

```python
df_sucio.describe()['seguidores']
```

```python
for i in random.sample(range(len(df_sucio)),5):
    df_sucio.loc[i,'correo'] = 'correo_invalido'
```
