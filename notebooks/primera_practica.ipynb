```python
import pandas as pd
```

```python
import json
import random
from datetime import datetime, timedelta

def generar_datos(cantidad):
    nombres = ["Juan", "María", "Pedro", "Ana", "Carlos", "Laura", "Diego", "Sofía", "Miguel", "Lucía"]
    ciudades = ["CDMX", "Monterrey", "Guadalajara", "Puebla", "Cancun", "Merida", "Tijuana", "Yucatan"]
    categoria = ["Tegnologia", "Ropa", "Hogar", "Videojuegos", "Deportes", ]
    usuarios = []

    # Fecha base para el cálculo de fecha_registro
    fecha_inicio = datetime(2020, 1, 1)

    print(f"Generando {cantidad} registros... Por favor, espera.")

    for i in range(cantidad):
        # Creamos cada objeto individual
        registro = {
            "usuario": random.choice(nombres),
            "edad": random.randint(18, 80),
            "ciudad": random.choice(ciudades),
            "compras_realizadas": random.randint(0, 50),
            "activo": random.choice([True, False]),
            "categoria": random.choice(categoria),
            "fecha_registro": (fecha_inicio + timedelta(days=random.randint(0, 1500))).strftime("%Y-%m-%d")
        }
        usuarios.append(registro)

        # Opcional: imprimir progreso cada 250,000 registros
        if (i + 1) % 250000 == 0:
            print(f"Progreso: {i + 1} registros creados.")

    return usuarios

def guardar_en_json(datos, DataFrame):
    print(f"Escribiendo datos en {DataFrame}...")
    with open(DataFrame, 'w', encoding='utf-8') as f:
        json.dump(datos, f, ensure_ascii=False, indent=4)
    print("¡Archivo generado con éxito!")

# --- Ejecución ---
NUM_REGISTROS = 1000000
ARCHIVO = "datos_usuarios.json"

datos_generados = generar_datos(NUM_REGISTROS)
guardar_en_json(datos_generados, ARCHIVO)
```

```python
df = pd.read_json("datos_usuarios.json");
print(df)
```
