```python
import pandas as pd 
import numpy as np 
import kagglehub

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset Iris
iris = load_iris(as_frame = True)
print(iris.DESCR)
```

```python
dfIris = iris.frame
```

```python
dfIris.rename(columns= { "target": "especia" } , inplace = True)
```

```python
dfIris["especia"] = dfIris["especia"].map({
    0 : "setosa",
    1 : "versicolor",
    2: "virginica",
})
```

```python
dfIris.head()
```

```python
dfIris.tail()
```

## Identificar variables predictoras y variables objetivo 

```python
# Variable predictoria 
X = dfIris.drop(columns=["especia"]) 
print ("Variables predictoras")
print (X.columns)
```

```python
# Variable objetivo 
y = dfIris["especia"]
print ("Variables objetivo")
print (y.name)
```

## Dividir los datos en la prueba y entrenamiento

**80%** para entrenamiento
**20%** para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.20,
    random_state = 42
)
```

```python
print ("X_train: ", X_train.shape)
print ("X_test: ", X_test.shape) 
print ("y_train: ", y_train.shape) 
print ("y_test: ", y_test.shape) 
```

## Escalado de variables

```python
scaler = StandardScaler()
```

```python
## Escalamos las variables
scaler = StandardScaler()
X_train_escalado = scaler.fit_transform(X_train)
X_test_escalado = scaler.transform(X_test)
```

```python
print("Sin escalar")
print(X_train)
print("Escalado")
print(X_train_escalado)
```

## Entrenamiento del modelo KNN
En esta prueba utilizaremos k = 3

```python
modelo = KNeighborsClassifier(n_neighbors= 3)
modelo.fit(
    X_train_escalado, y_train
)
```

## Realizamos predicciones con variables de prueba

```python
prediccion = modelo.predict(X_test_escalado)
print(prediccion[:10])
```

```python
## Comparacion entre valores reales y valores predecidos
comparacion = pd.DataFrame({
    "especie rear": y_test.values,
    "especie predicha": prediccion
})
comparacion.head(30)
```

```python
## Evaluacion del modelo
```

```python
accuracy = accuracy_score(
    y_test,
    prediccion
)
print(f"Accuracy/exactitud : {accuracy: 4f}")
```

```python
## Matriz de confusion
cm = confusion_matrix(y_test, prediccion)
cm_df = pd.DataFrame(cm, index = modelo.classes_,columns = modelo.classes_)
cm_df
```

## Clasification report
proporciona metricas que permiten evaluar el desempeño de cada. clase: precision, recall,f1-score, suport

```python
print(classification_report(y_test, prediccion))
```
